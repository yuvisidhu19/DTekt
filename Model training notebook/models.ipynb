```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
bert_preprocess = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4")
```

```python
# Bert layers
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessed_text = bert_preprocess(text_input)
outputs = bert_encoder(preprocessed_text)

# Neural network layers
l = tf.keras.layers.Dropout(0.1, name="dropout")(outputs['pooled_output'])
l = tf.keras.layers.Dense(1, activation='sigmoid', name="output")(l)

# Use inputs and outputs to construct a final model
model = tf.keras.Model(inputs=[text_input], outputs = [l])
```

```python
METRICS = [
      tf.keras.metrics.BinaryAccuracy(name='accuracy'),
      tf.keras.metrics.Precision(name='precision'),
      tf.keras.metrics.Recall(name='recall')
]

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=METRICS)
```

```python
from sklearn.model_selection import train_test_split
X = df['Reviews']
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
from flask import Flask, request
import pickle
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
from flask_ngrok import run_with_ngrok
import os
app = Flask(__name__)
run_with_ngrok(app)


@app.route('/', methods = ['POST', 'GET'])
def model():
  data = request.get_json()
  filename = 'new_model.sav'
  real_model = pickle.load(open(filename, 'rb'))
  filename = 'pos_model.sav'
  pos_model = pickle.load(open(filename, 'rb'))

  test = []
  with open('data.txt', encoding="utf8") as f:
      for line in f:
          test.append(line.strip())

  result1 = real_model.predict(test)
  real = 0
  fake = 0
  for i in result1:
    if i >= 0.5:
      real += 1
    else:
      fake += 1

  result2 = pos_model.predict(test)
  pos = 0
  neg = 0
  for i in result2:
    if i >= 0.5:
      pos += 1
    else:
      neg += 1

  fakeness = fake/(fake + real)
  negativeness = neg/(neg + pos)
  answer = (fakeness*0.8 + negativeness*0.2) / 2
  val = {'value': answer}
  return val


  #print(type(data['name']), flush=True)
  #return data
  # ans = 0.75
  # return {'val': ans}

if __name__ == '__main__':
  app.run()
```
